```python
outlier_counts = {}
for column in df.select_dtypes(include=['float64', 'int64']).columns:
        Q1 = df[column].quantile(0.25)
        Q3 = df[column].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        
        outliers = ((df[column] < lower_bound) | (df[column] > upper_bound)).sum()
        outlier_counts[column] = outliers
total = 0
for column, count in outlier_counts.items():
    total+=1
    print(f"Colone'{column}': {count} outliers")
print(total)
```

```python
import numpy as np
from sklearn.metrics import *
from sklearn.linear_model import *
from sklearn.neighbors import *
def metriques(name_model, y_test, y_pred):

    print("metriques de " + name_model)

    mae = mean_absolute_error(y_test, y_pred)
    mse = mean_squared_error(y_test, y_pred)
    rmse = np.sqrt(mse)
    r2 = r2_score(y_test, y_pred)
    mape = mean_absolute_percentage_error(y_test, y_pred)
    medae = median_absolute_error(y_test, y_pred)

    # Affichage
    print(f"MAE: {mae}")
    print(f"MSE: {mse}")
    print(f"RMSE: {rmse}")
    print(f"R²: {r2}")
    print(f"MAPE: {mape}")
    print(f"MedAE: {medae}")

    return r2, mae, mse, rmse, mape, medae
```

```python
r2_score_lr, mae_lr, mse_lr, rmse_lr, mape_lr, medae_lr = metriques("LinearRegression", Y_test, Y_pred_LR)
```
